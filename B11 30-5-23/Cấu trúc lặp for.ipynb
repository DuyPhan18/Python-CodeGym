## for kết hợp với hàm range()

### Trường hợp 1 tham số

```python
# in ra các số tự nhiên từ 0 đên 50
i = 0
while i <= 50:
    print(i)
    i += 1
```

```python
for i in range(51):
    print(i)
```

### Ví dụ: In các số lẻ từ 0 đến 100

```python
for i in range(101):
    if i % 2 != 0:
        print(i)
```

### Trường hợp 2 tham số

```python
for i in range(1500,2700):
    if i % 7 == 0:
        print(i)
```

### Ví dụ: In các số chia hết cho 7 từ a đến b

```python
a = 1500
b = 2700

for i in range(a, b + 1):
    if i % 7 == 0:
        print(i)
```

### Trường hợp 3 tham số

```python
#in ra số lẽ trong khoảng 101 đến 200
for i in range(101, 201, 2):
    print(i)
```

### Ví dụ: Biểu diễn cấp số cộng

```python
# in ra 10 số hạng đầu tiên của cấp số cộng
n = 10
d = 5

for n in range(n, n + 10 * d, d):
    print(n, end=" ")
```

## lặp qua một chuỗi

```python
course_name = "PythonFundamental"

for i in range(len(course_name)):
    print(course_name[i])
```

```python
course_name = "PythonFundamental"

for c in course_name:
    print(c)
```

### Ví dụ: Đếm số lượng nguyên âm trong chuỗi

```python
course_name = "PythonFundamental"
count = 0

for c in course_name:
#     if c in ("a", "o", "e", "u", "i"):
    if c in "euoai":
        count += 1
print(f"Tổng cộng có : {count} nguyên âm")
```
