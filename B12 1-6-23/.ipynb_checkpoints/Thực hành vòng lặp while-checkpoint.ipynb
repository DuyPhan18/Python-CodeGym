```python
'''
Khởi động 1: Viết chương trình in ra tất cả các số chia hết cho cả 5 và 7 trong khoảng từ 1500 đến 2700
'''
```

```python
i = 1500

while i <= 2700:
    if i % 5 == 0 and i % 7 == 0:
        print(i, end=" ")
    i+=1
```

```python
h = int(input("Please, input height of triagle: "))
print("__________________________________\n")

# i là cột ngang
i = h
#k,j là biến đếm trong hàng ngang
k = j = 0

#lặp cột
while 0 < i <= h:
    j = i
    #lặp giá trị trong hàng ngang
    while 0 < j <= i:
        print("*", end=" ")
        j-=1 
        while k < j :
            print("-", end=" ")
            j-=1
            break
    i-=1
    print("\n")
        
```

```python
'''
Bài 4: Viết chương trình nhập vào một số nguyên n và in ra các số từ 1 đến n với các trường hợp ngoại lệ như sau:
- Nếu số đó chia hết cho 3 thì in "py"
- Nếu số đó chia hết cho 7 thì in "thon"
- Nếu số đó chia hết cho cả 3 và 4 thì in "python"

Ví dụ: n = 25
1
2
py
4
5
py
thon
8
py
10
11
py
13
thon
py
16
17
py
19
20
python
22
23
py
25
'''
```

```python
n = int(input("Please, enter a number: "))

i = 1

while i <= n:
    if i % 3 == 0:
        print("py")
    elif i % 7 == 0:
        print("thon")
    elif i % 3 == 0 and i % 4 == 0:
        print("python")
    else:
        print(i)
    i+=1
```

```python
'''
Bài 5: Viết chương trình nhập vào một số và kiểm tra xem số đó có phải là số nguyên tố hay không
Note: Số nguyên tố là số chỉ chia hết cho 1 và chính nó
'''
```

```python
n = int(input("Please enter a number: "))

i = n - 1

if n < 2:
    print(f"{n} is not a prime number")
else:
    while 2 <= i < n:
        if n % i == 0:
            print(f"{n} is not a prime number")
            break
        i-=1
    else:            
        print(f"{n} is a prime number")
        
```

```python
'''
Bài 6: Viết chương trình mô phỏng trò chơi kéo búa bao theo luật chơi như sau:
Quy ước: 1 - búa, 2 - kéo, 3 - bao
Khi trò chơi bắt đầu, yêu cầu người chơi nhập 1 số từ bàn phím đại diện cho kéo, búa, bao
Chương trình tự random ra một lựa chọn từ 1 đến 3.
Nếu người chơi thua thì tiếp tục chơi cho đến khi thắng thì kết thúc chương trình.
Ví dụ:

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 3
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 2
Máy tính chọn: 1
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 2
Người chơi thắng.
'''
```

```python
c = ("Kéo", "Búa", "Bao")
r = ("Người chơi thắng", "Người chơi thua")
i = 1

#vòng lặp lại khi player chưa thắng sau 3 lần
while i > 0: 
    #gán lại k = 1 sau 3 lần vẫn thua
    k = 1
    
    #vòng lặp tăng giá trị k 
    while 0 < k <= len(c):
        player = int(input("Người chơi chọn (1 - kéo, 2 - búa, 3 - bao): "))
        
        #kiểm ttra nếu người dùng nhập lơn hơn 3
        if player > len(c):
            print("Không hợp lệ! Vui lòng chọn lại từ 1 đến 3")
            break
            
        # gán giá trị biến k cho computer và in ra    
        computer = k
        print(f"Máy chọn: {computer}")
        
        #kiểm tra điều kiện trò chơi
        if player == computer:
            print("Hoà!")
        if (player == 1 and computer == 2) or  (player == 2 and computer == 3) or (player == 3 and computer == 1):
            player = r[1]
            print(player)
        if (player == 1 and computer == 3) or (player == 2 and computer == 1) or (player == 3 and computer == 2):
            player = r[0]
            #thoát khỏi vòng lặp của k
            break   
        k += 1
    #kiểm tra player đã thắng chưa
    if player == r[0]: 
        print(player)
        #ngưng vòng lặp i
        break
    # nếu chưa thắng thì tiếp tục chơi lại    
    else:
        print("Chơi lại!")
        print("____________________________________________\n")
```
