### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

````python
fried_chicken = 35000
hamburger = 50000
pepsi = 10000

print("-----MENU-----")
print("Fried chicken: 35 000")
print("Hamburger: 50 000")
print("Pepsi: 10 000")
print("--------------")

quantity_fc = int(input("Entter quantity of fried chicken: "))
fc_price = fried_chicken * quantity_fc
quantity_ham = int(input("Entter quantity of hamburger: "))
ham_price = hamburger * quantity_ham
quantity_ps = int(input("Entter quantity of pepsi: "))
ps_price = pepsi * quantity_ps
total_price = ( fc_price + ham_price + ps_price )
print(f"Tổng đơn hàng là : {total_price}")
if total_price > 4000000:
    discount_price = ((total_price*15)/100)
    print(f"Tổng tiền được giảm là: {discount_price}")
    total_price = total_price - discount_price
    print(f"Tổng tiền phải thanh toán là: {total_price}")
    print("Cảm ơn quý khách đã mua")
elif total_price >2000000:
    discount_price = ((total_price*10)/100)
    print(f"Tổng tiền được giảm là: {discount_price}")
    total_price = total_price - discount_price
    print(f"Tổng tiền phải thanh toán là: {total_price}")
    print("Cảm ơn quý khách đã mua")
else:
    print(f"Tổng tiền phải thanh toán là: {total_price}")
    print("Cảm ơn quý khách đã mua")
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III

````python
height = int(input("Enter your height(cm): "))
weight = int(input("Enter your weight(kg): "))
height = height/100
print(f"height(m): {height}")
BMI = round(weight/height**2,2)

if BMI < 18.5:
    if BMI < 16:
        print(f"BMI: {BMI}, Gầy cấp độ III")
    if 16 <= BMI < 17:
        print(f"BMI: {BMI}, Gầy cấp độ II")
    if 17 <= BMI < 18.5:
        print(f"BMI: {BMI}, Gầy cấp độ I")
elif BMI > 30:
    if BMI > 40:
        print(f"BMI: {BMI}, Béo phì cấp độ III")
    if 35 <= BMI < 40:
        print(f"BMI: {BMI}, Béo phì cấp độ II")
    if 30 <= BMI < 35:
        print(f"BMI: {BMI}, Béo phì cấp độ I")
elif 25 < BMI < 30:
    print(f"BMI: {BMI}, Thừa cân")
elif 18.5 <= BMI < 25:
    print(f"BMI: {BMI}, Bình thường")
````
