```python
class Books:
    def __init__(self, book_name, book_cate, price):
        self.book_name = book_name
        self.book_cate = book_cate
        self.price = price
    def __str__(self):
        return f"Name of Book: {self.book_name}, Category: {self.book_cate}, price: {self.price}"
    
```

```python
def Add_book():
    global bk_name
    bk_name = name.get()
    global bk_cate 
    bk_cate = cate.get()
    global bk_price 
    bk_price = price.get()
    
    if bk_name == "" or bk_cate =="" or bk_price =="":
        messagebox.showerror("Error!", "Please Enter name, cate and price!")
    else:
        new_book = Books(bk_name, bk_cate, int(bk_price))
        book_lst.append(new_book)
        messagebox.showinfo("Success!", f"Added {bk_name} successfully!")
        print([book.book_name for book in book_lst])
        name.delete(0,"end")
        cate.delete(0,"end")
        price.delete(0,"end")
        
        
def Del_book():
    delete_name = search_name.get()
    check = False
    for i in range(len(book_lst)):
        if book_lst[i].book_name == delete_name:
            del book_lst[i]
            print([book.book_name for book in book_lst])
            messagebox.showinfo("Delete success!", f"Deleted {delete_name} successfully!")
            check = True
            break
    if check == False:
        messagebox.showerror("Delete error!", f"Do not found {delete_name} !")

def Search_book():
    search_bk_name = search_name.get()
    check = False
    name.delete(0,"end")
    cate.delete(0,"end")
    price.delete(0,"end")
    for i in range(len(book_lst)):
        if book_lst[i].book_name == search_bk_name:
            name.insert(0, book_lst[i].book_name)
            cate.insert(0, book_lst[i].book_cate)
            price.insert(0, book_lst[i].price)
            btn_update.config(state = ACTIVE)
            return book_lst[i]
            check = True
    if check == False:
         messagebox.showerror("Seach error!", f"Do not found {search_bk_name} !")
            
def Update_book():
    search_bk_name = search_name.get()
    check = False
    name.delete(0,"end")
    cate.delete(0,"end")
    price.delete(0,"end")
    new_price = price.get()
    for i in range(len(book_lst)):
        if book_lst[i].book_name == search_bk_name:
            check = True
            print(type(new_price))
            book_lst[i].price = int(new_price)
            print(book_lst[i].price)
    print([book.price for book in book_lst])
    if check == False:
         messagebox.showerror("Seach error!", f"Do not found {search_bk_name} !")
        

    
            
            
```

```python
from tkinter import *
from tkinter import messagebox
app = Tk()
app.title("Book Store Managent")
app.geometry("800x600")
book_lst = []
#Label
lbl_name = Label(app, text = "Book Name")
lbl_cate = Label(app, text = "Book Category")
lbl_price = Label(app, text = "Price")

#Input
name = Entry(app)
cate = Entry(app)
price = Entry(app)
search_name = Entry(app)

#Button
btn_add = Button(app, text ="ADD NEW", command = Add_book)
btn_del = Button(app, text ="DELETE", command = Del_book)
btn_update = Button(app, text ="UPDATE", state = DISABLED, command = Update_book)
btn_search = Button(app, text ="SEARCH", command = Search_book )


search_name.place(x = 100, y = 50)
btn_search.place(x = 200, y = 50)

lbl_name.place(x = 100, y = 100)
name.place(x = 200, y = 100)
lbl_cate.place(x = 100, y = 150)
cate.place(x = 200, y = 150)
lbl_price.place(x = 100, y = 200)
price.place(x = 200, y = 200)

btn_add.place(x = 500, y = 100)
btn_del.place(x = 500, y = 150)
btn_update.place(x = 500, y = 200)




app.mainloop()
```
