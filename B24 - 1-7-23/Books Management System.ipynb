```python
class Books:
    def __init__(self, book_name, book_cate, price):
        self.book_name = book_name
        self.book_cate = book_cate
        self.price = price
```

```python
def Add_book():
    global bk_name
    bk_name = name.get()
    global bk_cate 
    bk_cate = cate_chosen.get()
    global bk_price 
    bk_price = price.get()
    name_lst = []
    if bk_name == "" or bk_cate =="" or bk_price =="":
        messagebox.showerror("Error!", "Please Enter name, cate and price!")
    else:
        if len(book_lst) != 0:
            for i in range(len(book_lst)):
                name_lst.append(book_lst[i].book_name)
            if bk_name not in name_lst:
                new_book = Books(bk_name, bk_cate, int(bk_price))
                book_lst.append(new_book)
                name.delete(0,"end")
                cate_chosen.current()
                price.delete(0,"end")
              
                for i in range(len(book_lst)):
                    data = book_lst[i].book_name+", "+book_lst[i].book_cate+", "+str(book_lst[i].price)
                    file.write(f"[{data}]")
                    file.write("\n")
                    tree.insert("", text="", values=(book_lst[i].book_name, book_lst[i].book_cate,book_lst[i]. price))
                file.close()
            else:
                messagebox.showerror("Duplicate!", f"{bk_name} is exists!")
        else:
            new_book = Books(bk_name, bk_cate, int(bk_price))
            book_lst.append(new_book)
            print([book.book_name for book in book_lst])
            name.delete(0,"end")
            cate_chosen.current(0)
            price.delete(0,"end")
    for i, item in enumerate(book_lst):
         tree.insert("", i, text="", values=(item.book_name, item.book_cate, item.price))
    
        
def Del_book():
    delete_name = search_name.get()
    check = False
    for i in range(len(book_lst)):
        if book_lst[i].book_name == delete_name:
            del book_lst[i]
            print([book.book_name for book in book_lst])
            name.delete(0,"end")
            cate_chosen.current()
            price.delete(0,"end")
            messagebox.showinfo("Delete success!", f"Deleted {delete_name} successfully!")
            check = True
            break
    if check == False:
        messagebox.showerror("Delete error!", f"Do not found {delete_name} !")

def Search_book():
    search_bk_name = search_name.get()
    check = False
    name.delete(0,"end")
    cate_chosen.current()
    price.delete(0,"end")
    for i in range(len(book_lst)):
        if book_lst[i].book_name == search_bk_name:
            name.insert(0, book_lst[i].book_name)
            cate_chosen.insert(0, book_lst[i].book_cate)
            price.insert(0, book_lst[i].price)
            btn_update.config(state = ACTIVE,bg="#6296b7")
            return book_lst[i]
            check = True
    if check == False:
         messagebox.showerror("Seach error!", f"Do not found {search_bk_name} !")
                
def Update_book():
    new_price = price.get()
    Search_book().price = int(new_price)
    for i, item in enumerate(book_lst):
         tree.insert("", i, text="", values=(item.book_name, item.book_cate, item.price))
    for i in range (len(book_lst)):print(book_lst[i])
            
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import mysql.connector
from tkinter import *


mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="duyphan18",
  database = "demo1"
)
cursor = mydb.cursor()

query = 'SELECT * FROM book'
query_1 = mydb.cursor()
query_1.execute(query)
rows = query_1.fetchall()


app = tk.Tk()
app.title("Book Store Managent")
app.geometry("800x600")
book_lst = []

file = open("demo.txt", "a")

#Label
lbl_name = Label(app, text = "Book Name")
lbl_cate = Label(app, text = "Book Category")
lbl_price = Label(app, text = "Price")

#Input
name = Entry(app, width=30)

n = tk.StringVar()
cate_chosen = ttk.Combobox(app, width = 27, textvariable = n)
  
# Adding combobox drop down list
cate_chosen['values'] = (' Textbook ', 
                          ' Comic',
                          ' Poem',
                          ' Thriller book',
                          ' Dictionary ',
                          ' Cookery book ',
                          ' Science fiction book',
                          ' Reference book',
                          ' Novel ',
                          ' Encyclopedia ',
                          ' Myth',
                          ' Ghost story')
  
price = Entry(app, width=30)
search_name = Entry(app, width=50)

#Button"
btn_add = Button(app, text ="ADD NEW", bd=0, bg="#6296b7", justify=LEFT, height = 2, width = 10, command = Add_book)
btn_del = Button(app, text ="DELETE", bd=0, bg="#ea6a64", justify=LEFT, height = 2, width = 10, command = Del_book)
btn_update = Button(app, text ="UPDATE", bd=0, bg="#c4c4c4", justify=LEFT, height = 2, width = 10, state = DISABLED, command = Update_book)
btn_search = Button(app, text ="SEARCH", bd=0, bg="#009900", height = 1, width = 10, command = Search_book )
#treeview
columns = ('book_name', 'book_cate', 'price')
tree = ttk.Treeview(app, columns=columns, show='headings')
# define headings
tree.heading('book_name', text='Book Name')
tree.heading('book_cate', text='Category')
tree.heading('price', text='Price')
#load db into treeview
for i in rows:
    tree.insert('','end',iid=i[0], value = i)
for i in range(1, 4):
    tree.column("#"+str(i), anchor=tk.CENTER)

#Place widget
lbl_name.place(x = 100, y = 50)
name.place(x = 200, y = 50)
lbl_cate.place(x = 100, y = 100)
cate_chosen.place(x =200, y = 100)
cate_chosen.current(0)
lbl_price.place(x = 100, y = 150)
price.place(x = 200, y = 150)

search_name.place(x = 100, y = 220)
btn_search.place(x = 400, y = 220)

btn_add.place(x = 620, y = 50)
btn_del.place(x = 620, y = 100)
btn_update.place(x = 620, y = 150)
tree.place(x=100, y=250)



app.mainloop()
```
