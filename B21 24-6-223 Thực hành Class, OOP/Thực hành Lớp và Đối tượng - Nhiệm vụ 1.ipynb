# Nhiệm vụ 1: Bài toán phân số

## Giới thiệu:
Trong nhiệm vụ này, chúng ta sẽ làm quen với cách sử dụng `class` cũng như các cú pháp cơ bản trong lập trình hướng đối tượng. Chúng ta sẽ phát triển một class `PhanSo` dùng để thực hiện các thao tác cơ bản và các phép tính đơn giản với Phân Số.

## Bài 1:  \__init\__
Bước đầu tiên để khởi tạo một `class` là xây dựng hàm `__init__` để mô tả các thuộc tính của `class`.

Bạn hãy hoàn thành đoạn code dưới đây để tạo ra một lớp Phân số gồm có 2 thuộc tính là `tu_so` bằng 1 và `mau_so` bằng 2. Sau đó tạo ra một đối tượng của lớp Phân số và hiển thị thông tin tương ứng.

````python
class PhanSo:
    def __init__(self, tu_so, mau_so):
        self.tu_so = tu_so
        self.mau_so = mau_so
````

````python
a = PhanSo(1, 2)
print(a.tu_so)  # in ra 1
print(a.mau_so)  # in ra 2
````

## Bài 3: \__str\__
Hàm `__str__()` là một hàm được sử dụng khá phổ biến trong một `class`. Hàm `str()` sẽ giúp chúng ta định nghĩa ra một định dạng hiển thị bằng chữ của một object trong class đó. Ví dụ:

Ta có class `Student` dưới đây:

````python
class Student:
    def __init__(self):
        self.name = "Micheal"
````

````python
student = Student()
````

Khi không có hàm `__str__()` thì ta sẽ có được kết quả như sau:

````python
str(student)
````

Để định nghĩa được định dạng bằng chữ của `Student`, ta thiết lập hàm `__str__()`.

````python
class Student:
    def __init__(self):
        self.name = "Micheal"
        
    def __str__(self):
        return f"Hi! My name is {self.name}"
````

Và kết quả chúng ta thu được là:

````python
student = Student()
str(student)
````

**Nhiệm vụ:** Hãy thêm một hàm `__str__()` vào class `PhanSo` để thiết lập định dạng string của class này theo format sau:

```

Mẫu số là: 3
```

````python
class PhanSo:
    def __init__(self,tu_so, mau_so):
        self.tu_so = tu_so
        self.mau_so = mau_so
    
    def __str__(self):
        return f" Tử số là : {self.tu_so} \n Mẫu số là : {self.mau_so}"
````

## Bài 4: Nhân phân số
Phép nhân là phép tính đơn giản nhất đối với phân số. Để thực hiện được phép nhân phân số, ta làm theo quy tắc sau:


**Nhiệm vụ:** Thêm một hàm `multiply()` vào class `PhanSo` để thực hiện phép nhân phân số. Hàm này sẽ có một parameter `other` đại diện cho một object phân số khác. Hàm `multiply()` sẽ thực hiện nhân 2 phân số `self` và `other` lại với nhau và trả về một object `PhanSo` mới chứa tử số và mẫu số là kết quả của phép nhân vừa rồi.

````python
class PhanSo:
    def __init__(self,tu_so, mau_so):
        self.tu_so = tu_so
        self.mau_so = mau_so
    
    def __str__(self):
        return f" Tử số là : {self.tu_so} \n Mẫu số là : {self.mau_so}"
    
    def multiply(self, other):
        a = self.tu_so * other.tu_so
        b = self.mau_so * other.mau_so
        return PhanSo(a, b)
````

````python
a = PhanSo(4, 5)
b = PhanSo(15, 20)

result = a.multiply(b)

print(str(result))  
# Tử số là: 60
# Mẫu số là: 100
````

## Bài 5: Chia phân số

````python
class PhanSo:
    def __init__(self,tu_so, mau_so):
        self.tu_so = tu_so
        self.mau_so = mau_so
    
    def __str__(self):
        return f" Tử số là : {self.tu_so} \n Mẫu số là : {self.mau_so}"
    
    def multiply(self, other):
        a = self.tu_so * other.tu_so
        b = self.mau_so * other.mau_so
        return PhanSo(a, b)
    def divide(self, other):
        c = self.tu_so * other.mau_so
        d = self.mau_so * other.tu_so
        return PhanSo(c, d)
````

````python
a = PhanSo(4, 5)
b = PhanSo(15, 20)

result = a.divide(b)

print(str(result))  
# Tử số là: 80
# Mẫu số là: 75
````

## Bài 6: Cộng trừ
**Nhiệm vụ:** 2 phép tính cuối cùng là cộng và trừ phân số. Hãy viết các hàm `add()` và `subtract()` để thực hiện phép cộng và phép trừ 2 phân số.


````python
class PhanSo:
    def __init__(self,tu_so, mau_so):
        self.tu_so = tu_so
        self.mau_so = mau_so
    
    def __str__(self):
        return f" Tử số là : {self.tu_so} \n Mẫu số là : {self.mau_so}"
    
    def multiply(self, other):
        a = self.tu_so * other.tu_so
        b = self.mau_so * other.mau_so
        return PhanSo(a, b)
    def divide(self, other):
        c = self.tu_so * other.mau_so
        d = self.mau_so * other.tu_so
        return PhanSo(c, d)
    def add(self, other):
        e = (self.tu_so * other.mau_so) + (other.tu_so * self.mau_so)
        f = self.mau_so * other.mau_so
        return PhanSo(e, f)
    def subtract(self, other):
        h = (self.tu_so * other.mau_so) - (other.tu_so * self.mau_so)
        i = self.mau_so * other.mau_so
        return PhanSo(h, i)
        
````

````python
a = PhanSo(4, 6)
b = PhanSo(6, 6)

add_result = a.add(b)

subtract_result = a.subtract(b)

print(str(add_result))  
# Tử số là: 60
# Mẫu số là: 36

print(str(subtract_result))  
# Tử số là: -12
# Mẫu số là: 36
````

## Bài 7: Phân số tối giản
Chúng ta đã hoàn thành các phép tính cơ bản với phân số bằng cách thiết lập các thuộc tính và hàm trong class `PhanSo`. Hãy thiết lập hàm `minimalize()` để thực hiện tối giản phân số.

**Gợi ý:** hãy xây dựng một hàm để tìm ước chung lớn nhất giữa tử số và mẫu số để phục vụ cho việc tối giản nhé.

````python
import math
class PhanSo:
    def __init__(self,tu_so, mau_so):
        self.tu_so = tu_so
        self.mau_so = mau_so
    
    def __str__(self):
        return f" Tử số là : {self.tu_so} \n Mẫu số là : {self.mau_so}"
    
    def multiply(self, other):
        a = self.tu_so * other.tu_so
        b = self.mau_so * other.mau_so
        return PhanSo(a, b)
    def divide(self, other):
        c = self.tu_so * other.mau_so
        d = self.mau_so * other.tu_so
        return PhanSo(c, d)
    def add(self, other):
        e = (self.tu_so * other.mau_so) + (other.tu_so * self.mau_so)
        f = self.mau_so * other.mau_so
        return PhanSo(e, f)
    def subtract(self, other):
        h = (self.tu_so * other.mau_so) - (other.tu_so * self.mau_so)
        i = self.mau_so * other.mau_so
        return PhanSo(h, i)
    def minimalize(self):
        s = math.gcd(self.tu_so, self.mau_so)
        j = self.tu_so // s 
        k = self.mau_so // s 
        return PhanSo(j, k)
        
````

````python
a = PhanSo(4, 6)
b = PhanSo(6, 6)

add_result = a.add(b)

print(str(add_result))  
# Tử số là: 60
# Mẫu số là: 36

minimalize_result = add_result.minimalize()

print(str(minimalize_result))  
# Tử số là: 5
# Mẫu số là: 3
````
