## Câu lệnh if

### Thực hành:
Viết chương trình nhập vào môn học hiện tại. Nếu môn học là python thì in ra "I'm studying python"

```python
subject = input("Input your subject: ")
if subject.lower() == "python":
    print("I'm studying python")
```

## Câu lệnh if..else

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("even number")
else:
    print("odd number")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.

```python
my_str = input("Enter a string: ")
if my_str.count("python")>0:
    print("yes")
else: 
    print("no")
```

## Câu lệnh if…elif…else

### Thực hành: 
Viết chương trình nhập vào số tháng trong năm và cho biết tháng đó có bao nhiêu ngày

```python
month = int(input("Enter a month: "))
if month == 2: print(28)
elif month % 2 != 0: print(31)
elif month % 2 == 0: print(30)
```
