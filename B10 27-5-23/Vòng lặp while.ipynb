## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
n = int(input("Input n: "))

i = 0

while i <= n: #thiết lập điều kiện
    if i % 2 == 0:
#     if not i % 2:
        print(i)
    i+=1    # Cập nhật lại giá trị
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
n = int(input("Nhập số n: "))

i = n

while i >= 1:
    if i % 2 != 0:
        print(i)
    i-=1
```

### Ví dụ: Đảo ngược chuỗi

```python
my_str = input("enter a string: ")

i = len(my_str) - 1

while i  > 0:
    print(my_str[i], end="")    
    i-=1
```

### Ví dụ:

Lựa chọn k sao cho S(k) > a với k nhỏ nhất

In S(k) và k lên màn hình

```python
a = int(input("enter a: "))

k = 0
s = 0

while s <= a:
        k+=1
        s+=k
print(f"s là {s} tại k là {k}")
    
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

## Lệnh continue

```python
i = 100

while i <= 200:
    if i % 3 == 0 and i % 7 == 0:
        print( i )
    i+=1
```

```python
i = 100

while i <= 200:
    if i % 3 == 0 and i % 7 == 0:
        if i == 168:
            continue
        print( i )
        
    i+=1
```
