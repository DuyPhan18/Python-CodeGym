## Lp (class)

Mt lp l mt kiu d liu bao gm cc thuc tnh v cc phng thc c nh ngha t trc. y l s tru tng ha ca i tng. Khc vi kiu d liu thng thng, mt lp l mt n v (tru tng) bao gm s kt hp gia cc phng thc v cc thuc tnh. Hiu nm na hn l cc i tng c cc c tnh tng t nhau c gom li thnh mt lp i tng.

### Khai bo lp

`class <tn lp>:
    <khai bo bin>
    <khai bo hm>`
    
V d:

`class giangvien:
    ten = "Lm c Long"
    mon = "Python cn bn"
    def inten(self):
        print(self.ten)        
dien = giangvien()
print(dien.mon)
dien.inten()`

##### Lu :

- giangvien l 1 lp
- dien l 1 i tng c tt c thuc tnh (c bin ten, mon v hm inten) ca lp giangvien
- inten l 1 hm trong lp giangvien
- dien.mon l 1 bin trong i tng dien
- dien.inten() l cch  gi hm inten trong i tng dien


<b> Hy chy th on lnh  v d: </b>

```python
class giangvien:     
    ten = "Lm c Long"     
    mon = "Python cn bn"     
    def inten(self):         
        print(self.ten)         
giangvien = giangvien() 
print(giangvien.mon) 
giangvien.inten()
```

## i tng
Mt i tng bao gm 2 thng tin: **thuc tnh** v **phng thc**.


**Thuc tnh** chnh l nhng thng tin, c im ca i tng. V d: con ngi c cc c tnh nh mt, mi, tay, chn

**Phng thc** l nhng thao tc, hnh ng m i tng  c th thc hin. V d: mt ngi s c th thc hin hnh ng ni, i, n, ung, . . .

```python
#to class vehicle dng vai tr l 1 skeleton cho ptien giao thong ng b
#mi 1 ptgt c cc thuc tnh: model, brand, color
#mt ptgt c hnh dng move_forward(). khi gi hm s print("I'm moving forward!")

class vehicle:
    model = "SUV"
    brand = "Toyota"
    color = "Red"
    def move_forward(self):
        print("I'm moving forward!")
    def intro(self):
        print(f"Model: {self.model}")
        print(f"Brand: {self.brand}")
        print(f"Color: {self.color}")
        self.move_forward()
xe = vehicle()
xe.intro()
```

## T kha self

## Hm init

```python
class vehicle:
    def __init__(self, model, brand, color):
        self.model = model
        self.brand = brand
        self.color = color
    def move_forward(self):
        print("I'm moving forward!")
    def intro(self):
        print(f"Model: {self.model}")
        print(f"Brand: {self.brand}")
        print(f"Color: {self.color}")
        self.move_forward()
xe = vehicle("Sedan", "Toyota", "Red")
xe.intro()
```

```python
xe2 = vehicle("SUV", "Mazda", "Silver")
xe2.intro()
```

## Thc hnh:

```python
# M t mt class Rectangle gm cc thuc tnh chiu di v rng. 
# Pht trin cc hm tnh chu vi v din tch 
# To ra 2 object vi cc kch thc khc nhau
```

```python
class rectangle:
    def __init__(self, d, r):
        self.d = d
        self.r = r
    def tinh_cv(self):
        return (self.d + self.r ) * 2
    def tinh_dt(self):
        return self.d * self.r
    def show(self):
        print(f"Hinh chu nhat c d, r la: {self.d}, {self.r}")
        
rec_1 = rectangle(18,7)
rec_1.show()
print(f"Chu vi hinh chu nhat 1 la:{rec_1.tinh_cv()}")
print(f"Dien tich hinh chu nhat 1 la: {rec_1.tinh_dt()}")
print("")
rec_2 = rectangle(11,4)
rec_2.show()
print(f"Chu vi hinh chu nhat 2 la:{rec_2.tinh_cv()}")
print(f"Dien tich hinh chu nhat 2 la: {rec_2.tinh_dt()}")
```
