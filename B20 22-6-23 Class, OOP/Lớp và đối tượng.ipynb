# Object Oriented Programming

**Khái niệm:** Lập trình hướng đối tượng (OOP) là một kỹ thuật lập trình cho phép lập trình viên tạo ra các đối tượng trong code trừu tượng hóa các đối tượng.

## Lớp (class)

Một lớp là một kiểu dữ liệu bao gồm các thuộc tính và các phương thức được định nghĩa từ trước. Đây là sự trừu tượng hóa của đối tượng. Khác với kiểu dữ liệu thông thường, một lớp là một đơn vị (trừu tượng) bao gồm sự kết hợp giữa các phương thức và các thuộc tính. Hiểu nôm na hơn là các đối tượng có các đặc tính tương tự nhau được gom lại thành một lớp đối tượng.

### Khai báo lớp

`class <tên lớp>:
    <khai báo biến>
    <khai báo hàm>`
    
Ví dụ:

`class giangvien:
    ten = "Lâm Đức Long"
    mon = "Python căn bản"
    def inten(self):
        print(self.ten)        
dien = giangvien()
print(dien.mon)
dien.inten()`

##### Lưu ý:

- giangvien là 1 lớp
- dien là 1 đối tượng có tất cả thuộc tính (có biến ten, mon và hàm inten) của lớp giangvien
- inten là 1 hàm trong lớp giangvien
- dien.mon là 1 biến trong đối tượng dien
- dien.inten() là cách để gọi hàm inten trong đối tượng dien


<b> Hãy chạy thử đoạn lệnh ở ví dụ: </b>

```python
class giangvien:     
    ten = "Lâm Đức Long"     
    mon = "Python căn bản"     
    def inten(self):         
        print(self.ten)         
giangvien = giangvien() 
print(giangvien.mon) 
giangvien.inten()
```

## Đối tượng
Một đối tượng bao gồm 2 thông tin: **thuộc tính** và **phương thức**.


**Thuộc tính** chính là những thông tin, đặc điểm của đối tượng. Ví dụ: con người có các đặc tính như mắt, mũi, tay, chân…

**Phương thức** là những thao tác, hành động mà đối tượng đó có thể thực hiện. Ví dụ: một người sẽ có thể thực hiện hành động nói, đi, ăn, uống, . . .

```python
#tạo class vehicle dóng vai trò là 1 skeleton cho ptien giao thong đường bộ
#mỗi 1 ptgt có các thuộc tính: model, brand, color
#một ptgt có hành dộng move_forward(). khi gọi hàm sẽ print("I'm moving forward!")

class vehicle:
    model = "SUV"
    brand = "Toyota"
    color = "Red"
    def move_forward(self):
        print("I'm moving forward!")
    def intro(self):
        print(f"Model: {self.model}")
        print(f"Brand: {self.brand}")
        print(f"Color: {self.color}")
        self.move_forward()
xe = vehicle()
xe.intro()
```

## Hàm init

```python
class vehicle:
    def __init__(self, model, brand, color):
        self.model = model
        self.brand = brand
        self.color = color
    def move_forward(self):
        print("I'm moving forward!")
    def intro(self):
        print(f"Model: {self.model}")
        print(f"Brand: {self.brand}")
        print(f"Color: {self.color}")
        self.move_forward()
xe = vehicle("Sedan", "Toyota", "Red")
xe.intro()
```

```python
xe2 = vehicle("SUV", "Mazda", "Silver")
xe2.intro()
```

## Thực hành:

```python
# Mô tả một class Rectangle gồm các thuộc tính chiều dài và rộng. 
# Phát triển các hàm tính chu vi và diện tích 
# Tạo ra 2 object với các kích thước khác nhau
```

```python
class rectangle:
    def __init__(self, d, r):
        self.d = d
        self.r = r
    def tinh_cv(self):
        return (self.d + self.r ) * 2
    def tinh_dt(self):
        return self.d * self.r
    def show(self):
        print(f"Hinh chu nhat có d, r la: {self.d}, {self.r}")
        
rec_1 = rectangle(18,7)
rec_1.show()
print(f"Chu vi hinh chu nhat 1 la:{rec_1.tinh_cv()}")
print(f"Dien tich hinh chu nhat 1 la: {rec_1.tinh_dt()}")
print("")
rec_2 = rectangle(11,4)
rec_2.show()
print(f"Chu vi hinh chu nhat 2 la:{rec_2.tinh_cv()}")
print(f"Dien tich hinh chu nhat 2 la: {rec_2.tinh_dt()}")
```
